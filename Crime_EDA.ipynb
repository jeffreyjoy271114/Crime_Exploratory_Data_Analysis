## Answer some of the questions

### 1.What are the most common crimes in terms of offence groups?

```python
# Getting the counts of each offense code group
crime.OFFENSE_CODE_GROUP.value_counts()
```

```python
# Creating the percentage dataframe of the top 10 offense group

offense_group_vals = crime.OFFENSE_CODE_GROUP.value_counts()[:10]

display(offense_group_vals / crime.shape[0])

# Creating a bar plot of thr top 10 offence groups
(offense_group_vals / crime.shape[0]).plot(kind = 'bar');
plt.title("Top 10 Offense Group in Percentage(%) of all crimes")
```

### 2. What are the Least common Offense groups?


```python
crime.OFFENSE_CODE_GROUP.value_counts()
```

```python
# Creating the bottom or the least 10 offense group dataframe  

offense_group_vals = crime.OFFENSE_CODE_GROUP.value_counts(ascending=True)[:10]

display(offense_group_vals / crime.shape[0])

# Creating a bar plot of thr top 10 offence groups
(offense_group_vals / crime.shape[0]).plot(kind = 'bar');
plt.title("Bottom or Least 10 Offense Group in Percentage(%) of all crimes")
```

### 8. On what Days and during which hours are the most crimes commited?

```python
# Creating a variable in which the groupby function is used to group the hour and the day of the week to get on what 
# what Days and during which hours are the most crimes commited?

week_and_hour = crime.groupby(['HOUR', 'DAY_OF_WEEK']).count()['INCIDENT_NUMBER'].unstack()
```

```python
# Rearanging the columns 
week_and_hour.columns = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
```

```python
# Visualizing it using the heatmap via seaborn library
sns.heatmap(week_and_hour, cmap = sns.cubehelix_palette(as_cmap=True));
```

### 9. In which months were the number of crimes below average?
### 10. In which months on average did the most crimes occur?

```python
# If the Value is less than the Average we will highlight it with blue.

avg_crime = crime.groupby(['YEAR', 'MONTH']).count()['INCIDENT_NUMBER'].mean()
print(f'The Avereage number of Crimes is {avg_crime}.')

year_month = crime.groupby(['MONTH', 'YEAR']).count()['INCIDENT_NUMBER'].unstack()

def style_negative(v,props=''):
                   return props if v < avg_crime else None

s2 = year_month.style.applymap(style_negative, props='color:blue;')\
                   .applymap(lambda v:'opacity:20%;' if (v<30) and (v>-0.3) else None)
s2
                   
                   
```

```python
# Using the apply function to highlight the maximum in a column in darkgreen

def highlight_max(s, props = ''):
    return np.where(s == np.nanmax(s.values), props, '')
s2.apply(highlight_max, props = 'color:white; background-color:darkgreen', axis = 0)
                    
```
